```python
from transformers import pipeline
import matplotlib.pyplot as plt

toxicity_analyzer = pipeline(
    "text-classification",
    model="unitary/toxic-bert",
    device="cuda"
)

text = "Racial discriminators carry out massacres against specific races"

result = toxicity_analyzer(text)

res = result[0]['score'] * 100.0

if res > 0.5:
  tox = "TOXIC"
else:
  tox = "SAFE"

print(f"input: {text}")
print(f"score: {res:.8f}%")
print(f"result: {tox}")
```
